```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[2,4,6]
plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y,color="black",marker="o")
plt.show()
```

```python
#x^3
import numpy as np
x=np.array([1,2,3,4])
y=x**3
a=np.arange(0,5,0.1)
b=a**3
plt.plot(a,b,color="blue",linewidth=3)
plt.plot(x,y,color="red",linewidth=3)
plt.show()
```

```python
#x^3 with labeling
import numpy as np
x=np.array([1,2,3,4])
y=x**3
a=np.arange(0,5,0.1)
b=a**3
plt.plot(a,b,color="blue",linewidth=3,label="curvy x^3")
plt.plot(x,y,color="red",linewidth=3,label="line mated")
plt.legend()                          #for label on graph
plt.text(2,80,"text",fontsize=10)     #for writing text on graph
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("x cube graph")
plt.axis([0,8,0,200])                 #for changing view range of graph
plt.grid()
plt.show()


```

```python
def abc(a):
  a=f"{a}%"
  return a

sizes=[2,3,5,6]
colors=["red","blue","pink","yellow"]
labels=["A","B","C","D"]
explode=[0.2,0,0,0]
plt.title("classes")
plt.axis("equal")
plt.pie(sizes,colors=colors,labels=labels,explode=explode,autopct=abc,counterclock=True,startangle=90)
plt.show()
```
